```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def dev(a,b):
    return a/b
def cal(a,b,o):
    if o=="+":
        print(add(a,b))
    elif o=="-":
        print(sub(a,b))
    elif o=="*":
        print(mul(a,b))
    elif (0=="/"):
        print(dev(a,b))
    elif (b==0):
        print("the value of b can not be zero.")
    else:
        print("invalid operator")
a=int(input("enter a="))
b=int(input("enter b="))
o=input("enter + or- or * or /=  ")
cal(a,b,o)
```
